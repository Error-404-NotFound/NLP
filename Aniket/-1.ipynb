```python
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
print(iris.DESCR)
```

```python
X=iris.data
```

```python
Y = iris.target
```

```python
Y
```

```python
Y.shape
```

```python
#one hot encoding
# class 0 --> [1,0,0,...]
# class 1 --> [0,1,0,...]
# class 2 --> [0,0,1,...]
```

```python
import tensorflow
```

```python
from keras.utils import to_categorical
```

```python
Y=to_categorical(Y)
```

```python
Y.shape
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar_object=MinMaxScaler()
```

```python
scalar_object.fit(X_train)
```

```python
scaled_X_train=scalar_object.transform(X_train)
```

```python
scaled_X_test=scalar_object.transform(X_test)
```

```python
scaled_X_train
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(8,input_dim=4,activation="relu"))
model.add(Dense(8,input_dim=4,activation="relu"))
model.add(Dense(3,activation="softmax"))
model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
model.summary()
```

```python
model.fit(scaled_X_train,y_train,epochs=500,verbose=1)
```

```python
scaled_X_test
```

```python
model.predict(scaled_X_test)
```

```python
model.predict_classes(scaled_X_test)
```

```python
predict_x=model.predict(scaled_X_test) 
classes_x=np.argmax(predict_x,axis=1)
```

```python
predictions = np.argmax(model.predict(scaled_X_test),axis=1)
```

```python
predictions
```

```python
y_test.argmax(axis=1)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
confusion_matrix(y_test.argmax(axis=1),predictions)
```

```python
print(classification_report(y_test.argmax(axis=1),predictions))
```

```python
accuracy_score(y_test.argmax(axis=1),predictions)
```
